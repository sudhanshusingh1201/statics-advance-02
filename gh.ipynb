**1.What is hypothesis testing in statistics
**

Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves making an initial assumption (called the null hypothesis) and then determining whether there's enough evidence in the sample to reject that assumption in favor of an alternative hypothesis.

**2.A What is the null hypothesis, and how does it differ from the alternative hypothesis?**

Null hypothesis (H₀): This is the default assumption that there is no effect or no difference. For example, “There is no difference in average weight between two groups.”

Alternative hypothesis (H₁ or Ha): This contradicts the null and represents the conclusion you want to test. For example, “There is a difference in average weight between the groups.”

**3.What is the significance level in hypothesis testing, and why is it important,**

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It defines how much risk you're willing to accept for a false positive. Common values are 0.05 (5%) or 0.01 (1%). A smaller α means stricter criteria for rejecting H₀.



**4. What does a P-value represent in hypothesis testing,**

A P-value measures the probability of obtaining test results at least as extreme as the observed data, assuming the null hypothesis is true. It helps you decide whether to reject H₀:

Low p-value (≤ α): Reject H₀.

High p-value (> α): Fail to reject H₀.



**5.How do you interpret the P-value in hypothesis testing,**

P ≤ 0.05 → Strong evidence against H₀ → Reject H₀.

P > 0.05 → Weak evidence against H₀ → Fail to reject H₀. A p-value does not measure the probability that H₀ is true — it only tells you how unusual your results are under H₀.



**6.What are Type 1 and Type 2 errors in hypothesis testing,**

Type I Error (False Positive): Rejecting H₀ when it's actually true. Risk = α.

Type II Error (False Negative): Failing to reject H₀ when it's actually false. Risk = β. There’s a trade-off between the two — reducing one can increase the other.



**7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing,**

One-tailed test: Tests for an effect in one direction (e.g., H₁: μ > 100).

Two-tailed test: Tests for an effect in either direction (e.g., H₁: μ ≠ 100).

Two-tailed tests are more conservative, requiring stronger evidence to reject H₀.



**8.A What is the Z-test, and when is it used in hypothesis testing,**

 Z-test is used when:

The population standard deviation is known.

The sample size is large (n > 30). It compares the sample mean to the population mean using the standard normal distribution (Z-distribution).



**9.How do you calculate the Z-score, and what does it represent in hypothesis testing,**

Z = (X̄ - μ) / (σ / √n)
Where:

X̄ = sample mean

μ = population mean

σ = population standard deviation

n = sample size

**10.A What is the T-distribution, and when should it be used instead of the normal distribution,**

The T-distribution is used when:

The sample size is small (n < 30).

The population standard deviation is unknown. It is similar to the normal distribution but has heavier tails (more variability), which accounts for the extra uncertainty.



**11.A What is the difference between a Z-test and a T-test,**

Z-test: Used when population standard deviation is known and sample size is large.

T-test: Used when population standard deviation is unknown and/or sample size is small. Both test for differences in means but use different distributions.



**12.What is the T-test, and how is it used in hypothesis testing,**

A T-test compares the means of one or two groups to assess if they are statistically different. Types:

One-sample t-test: Compare sample mean to a known value.

Two-sample t-test: Compare means of two independent groups.

Paired t-test: Compare means from the same group at different times.

**13.What is the relationship between Z-test and T-test in hypothesis testing,**

Both are used for comparing means:

Z-test assumes known σ, T-test does not.

As sample size increases, T-distribution approaches normal distribution, making T-test results close to Z-test results.

**14.What is a confidence interval, and how is it used to interpret statistical results,**

A confidence interval (CI) provides a range of values, derived from sample data, that is likely to contain the true population parameter.
Example: A 95% CI means we are 95% confident that the true mean lies within this interval.



**15.What is the margin of error, and how does it affect the confidence interval,**

The margin of error (MoE) is the amount added/subtracted from the sample estimate to create the CI.
Larger MoE → wider CI → more uncertainty.
MoE depends on:

Sample size

Confidence level

Standard deviation



**16.How is Bayes' Theorem used in statistics, and what is its significance,**

Bayes' Theorem updates the probability of a hypothesis based on new evidence. It combines prior probability with likelihood of observed data.
Used in:

Medical testing

Machine learning

Decision making under uncertainty



**17.A What is the Chi-square distribution, and when is it used,**

The Chi-square (χ²) distribution is used in categorical data analysis. It’s the distribution of a sum of squared standard normal variables.
Used for:

Testing independence in contingency tables

Goodness-of-fit tests



**18.A What is the Chi-square goodness of fit test, and how is it applied,**

This test checks whether observed frequencies match expected frequencies for a categorical variable.
Steps:

Define null hypothesis (e.g., “Distribution fits expected proportions”).

Calculate expected frequencies.

Compute χ² statistic.

Compare with critical value or use p-value.



**19.A What is the F-distribution, and when is it used in hypothesis testing,**

The F-distribution is used to compare variances of two populations. It arises in ANOVA and F-tests.
It’s asymmetrical and only positive, since it involves ratios of variances.



**20.What is an ANOVA test, and what are its assumptions,**

ANOVA (Analysis of Variance) tests if 3 or more group means are significantly different. Assumptions:

Independence of observations

Normality in each group

Equal variances (homogeneity of variance)



**21.What are the different types of ANOVA tests,**

One-way ANOVA: One independent variable with multiple groups.

Two-way ANOVA: Two independent variables.

Repeated Measures ANOVA: Same subjects tested under different conditions/times.



**22.What is the F-test, and how does it relate to hypothesis testing?**

An F-test compares two variances or multiple means (via ANOVA).
Used to:

Test variance equality

Analyze variance in multiple groups (in ANOVA) F = (variance between groups) / (variance within groups)

**1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the resultsC**

```python
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, n):
    z = (sample_mean - population_mean) / (population_std / (n ** 0.5))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

z, p = z_test(102, 100, 10, 50)
print(f"Z-score: {z:.2f}, P-value: {p:.4f}")
```

**5.C Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testingC**


```python
def visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    se = sigma / np.sqrt(n)
    x = np.linspace(-4, 4, 1000)
    null = norm.pdf(x, mu0, se)
    alt = norm.pdf(x, mu1, se)

    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    plt.plot(x, null, label='Null (H0)', color='blue')
    plt.plot(x, alt, label='Alternative (H1)', color='green')
    plt.fill_between(x, null, where=x > x_crit, color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, alt, where=x < x_crit, color='orange', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.grid()
    plt.title("Type I and II Errors")
    plt.show()

visualize_errors()
```

**9.Write a Python program to calculate the margin of error for a given confidence level using sample dataC**

```python
def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    t_crit = stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    return t_crit * sem

data = np.random.normal(80, 12, 40)
moe = margin_of_error(data)
print(f"Margin of Error: ±{moe:.2f}")
```

**11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the processC**

```python
def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    posterior = (prob_B_given_A * prior_A) / prob_B
    return posterior

# Example: Disease test
prior = 0.01      # P(Disease)
sensitivity = 0.99  # P(Positive | Disease)
false_positive = 0.05  # P(Positive | No Disease)
prob_B = sensitivity * prior + false_positive * (1 - prior)

posterior = bayes_theorem(prior, sensitivity, prob_B)
print(f"Posterior probability (Disease | Positive): {posterior:.4f}")
```

**17.+ Implement an F-test using Python to compare the variances of two random samples+**

```python
from scipy.stats import f

def f_test(data1, data2):
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
    f_stat = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
    p_value = 1 - f.cdf(f_stat, dfn, dfd)
    return f_stat, p_value

x = np.random.normal(0, 2, 30)
y = np.random.normal(0, 1, 30)
f_stat, p = f_test(x, y)
print(f"F-test => F-statistic: {f_stat:.2f}, P-value: {p:.4f}")
```

**20.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA**

```python
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # Normality
    for i, g in enumerate(groups):
        stat, p = shapiro(g)
        print(f"Group {i+1} normality p = {p:.4f}")
    
    # Homogeneity of variances
    stat, p = levene(*groups)
    print(f"Levene’s Test (equal variance) p = {p:.4f}")
```

```python
sample = np.random.normal(0, 5, 30)
sample_var = np.var(sample, ddof=1)
n = len(sample)
chi2_stat = (n - 1) * sample_var / 25  # population variance = 25

p = 1 - chi2.cdf(chi2_stat, df=n - 1)
print(f"Chi-square for variance => χ²: {chi2_stat:.2f}, P-value: {p:.4f}")
```
